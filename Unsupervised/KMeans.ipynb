# KMEANS  Example 

```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
#import seaborn as sns
#sns.set()
from sklearn.cluster import KMeans
from sklearn import preprocessing
################################################################################################
```

## Import data

```python
################################################################################################
# DATA
#------

data = pd.read_csv('kmeans.csv')
print("Data:",data)

# PLOT DATA
#----------

plt.scatter(data['Satisfaction'], data['Loyalty'])
plt.xlabel('Satisfaction')
plt.ylabel('Loyalty')
plt.show()

```

```python

################################################################################################
# CLUSTER DATA
#----------------
X = data.copy()
kmeans = KMeans(2)
kmeans.fit(X)

clusters = X.copy()
clusters['cluster_pred'] = kmeans.fit_predict(X)

# PLOT RESULTS
#----------------
plt.scatter(clusters['Satisfaction'], clusters['Loyalty'], c = clusters['cluster_pred'], cmap='rainbow')
plt.xlabel('Satisfaction')
plt.ylabel('Loyalty')
plt.show()

################################################################################################
#  STANDARDIZE DATA
#---------------------
X_scaled = preprocessing.scale(X)
clusters_new = X_scaled.copy()
# CLUSTER  DATA
#----------------
kmeans_new =  KMeans(4)
clusters['cluster_pred_new'] = kmeans_new.fit_predict(X_scaled)


# PLOT RESULTS
#----------------
plt.scatter(clusters['Satisfaction'], clusters['Loyalty'], c = clusters['cluster_pred_new'], cmap='rainbow')
plt.xlabel('Satisfaction')
plt.ylabel('Loyalty')
plt.show()
```

```python
################################################################################################
# ELBOW CURVE
########################

wcss = []
for i in range(1,30):
    kmeans = KMeans(i)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)


plt.plot(range(1,30), wcss)
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
################################################################################################

```
